### How Accessible is Long Distance Delivery and Can You Get Vegetarian Option Easily? 
(Cont.)

```python
#import the pandas library and aliasing as pd
#import the numpy library and aliasing as np
#import the matplotlib.pyplot library and aliasing as plt

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Swiggy_50.csv')

df3 = df[['Restaurant Name','Veg','city','Long Distance Delivery']]
df3 = df3[df3['city'] == 'Delhi'] ## narrows our focus just on restaurants in Delhi
df3 = df3.dropna() #The dropna() function is used to remove missing values

#USING GROUP BY FUNCTION
df3_group = df3.groupby('Long Distance Delivery')

#USING A QUERY TO FILTER 
df3_group_query=df3.query('Veg == True')

print(df3_group_query)
```

### How Accessible is Long Distance Delivery and Can You Get Vegetarian Option Easily? 
(Cont.) Sorting Long Distance Delivery 'Yes' option to top of list 

```python
#Sorting Long Distance Delivery 'Yes' option to top of list
df3_group_query_TABLE = df3_group_query.sort_values(by=['Long Distance Delivery'],ascending=False)
df3_group_query_TABLE 
```

#### INSIGHT 3:
