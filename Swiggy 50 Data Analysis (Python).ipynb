To begin import all modules and then load the CSV file into a new dataframe called DF.

```python
#import the pandas library and aliasing as pd
#import the numpy library and aliasing as np
#import the matplotlib.pyplot library and aliasing as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Swiggy_50.csv')
```

Dataframe 1 Cleaning (cont.): Is it More Expensive to go to a Nice Place?

```python
df = df[['Restaurant Name', 'Category', 'Rating', 'Cost for two', 'Veg', 'city']]
df = df[df['city'] == 'Delhi'] ## narrows our focus just on restaurants in Delhi
df = df.dropna() #The dropna() function is used to remove missing values
print(df.head(5))
print(df.tail(5))
```

Is it More Expensive to go to a Nice Place? (Scatter Plot)

```python
x = np.array(df['Cost for two'])
y = np.array(df['Rating'])
plt.figure(figsize=(15,7))
plt.scatter(x, y)
plt.title("Is it More Expensive to go to a Nice Place?", fontweight ="bold", color ="black")
plt.xlabel("Cost for Two in Rupees", fontweight ="bold", color ="black")
plt.ylabel("Rating on Scale 0 to 5", fontweight ="bold", color ="black")
plt.xticks(fontsize= 10, color ="black")
plt.yticks(fontsize= 10, color ="black")
plt.show()
```

Is it More Expensive to go to a Nice Place? (Sorting Demonstration - Top 5 vs Bottom 5 Data)

```python
df = df[['Restaurant Name','Cost for two','city']]
df = df[df['city'] == 'Delhi'] ## narrows our focus just on restaurants in Delhi
df = df.dropna() #The dropna() function is used to remove missing values
df.sort_values(by=['Cost for two'], ascending=False)
```
